```python
df=spark2.read.csv("telco_customer_churn.csv",header=True,inferSchema=True)
```

```python
df.dtypes
```

```python
L=df.dtypes
```

```python
L[0]
```

```python
for col in L:
    nC,dt=col
    if dt=='string':
        print("Nombre de la columna: ",nC)
        print("Tipo de dato: ",dt)
```

```python
var_cat=[nC for nC,dt in df.dtypes if dt=='string']    # Se queda con el nombre de la columna (nC) que son string
```

```python
var_cat
```

```python
var_num=[nC for nC,dt in df.dtypes if dt in ['int','double']]
```

```python
var_num
```

```python
var_cat.remove('TotalCharges')
var_num.append('TotalCharges')
```

```python
var_num
```

```python
df.select(var_num).describe().toPandas()
```
